We'll plot this below. First, a comment about **tildes**. We could use `plot( year, sst, ...)` but insead we'll use the `sst~year` notation, which indicates that the thing on the left (`sst`) is a response or a **dependent variable**, while the thing on the right (`year`) is one or more **independent variable(s)**.  This is also used in packages like `lattice`, `ggplot2` and `dplyr`.  More info at https://stackoverflow.com/a/14976479/269192 and details at https://stackoverflow.com/a/8055683/269192

```r
plot(sst~year, main="Historical sea surface temperature anomalies", ylab="Degrees Celsius",
  xlab="Year", type="b") # try plot(sst) vs. plot(sst~year), see what happens to the x-axis
abline(h=0)
# hmm, is there evidence of a trend?
```

#### Linear model for sst data: 

$$ \text{sst} = \beta_0 + \beta_1 \cdot \text{year} + \epsilon, \quad (\epsilon \; \text{iid} \sim \mathcal{N}(0,\sigma^2)) $$

We'll use `lm` from the `stats` package.

Later in this class we'll do more advanced linear regression where we account for the fact that the errors $\epsilon$ are very likely **not** independent.


```r
# The 'stats' package is pre-installed.
# We can list the libraries that are actually loaded doing
(.packages())
```

```r
model <- lm(sst~year) # automatically includes intercept... can remove via lm(sst~year-1)
model
print("And ask for a summary of the model")
summary(model) # note p-values, standard errors, r^2, etc.
```

```r
plot(sst~year, main="Historical sea surface temperature anomalies", ylab="Degrees",
  xlab="Year", type="b")

# Add lines
abline(h=0)

mean.fcn <- model$coef[1] + model$coef[2]*year
lines(mean.fcn~year)
```

#### Residual diagnostics
`par` (for "parameters") can be used to set or query graphical parameters

```r
par(mfrow=c(2,2)) # set up a 2x2 window for four plots
plot(model) # whoa, powerful!

# looking at the (1,1) plot, Residuals vs. Fitted, how is our assumption of iid errors?
dev.off() # manually kill plotting device (no more annotations allowed; do this for performance reasons sometimes)
```

```r
sstr <- sst - mean.fcn # residuals
qqnorm(sstr) ## normal q-q plot
qqline(sstr)
```

And look at the **autocorrelation** of the residuals (which we will learn about in this class)

```r
acf(sstr,lwd=3) # more later...
```
